9. What is an online learning system?
You split your data into groups and you feed the algorithm incrementally with these groups step by step

10. What is out-of-core learning?
It is a type of online learning use to train systems on huge datasets that can't fit in one machine's memory. The algorith chunks the data into pieces, runs a training step on that data and repeats the process until it has run on all of the data.

It's usually done offline. Think of it as incremental learning.

11. What type of learning algorithm relies on a similiraty measure to make predictions?

12. What is the difference between a model parameter and a learning algorithms hyperparameter?
Model parameters are the values the model learns to make accurate predictions, while hyperparameters are the settings you tweak to control the learning process.

13. What do model-based learning algorithms search for? What is the most common strategy they use to succeed? How do they make predictions?

15. If your model performs great on the training data but generalizes poorly to new instances, what is happening? Can you name three possible solutions?
This is a clear sign of overfitting.

You can simplify the model by selecting one with fewer parameters, gather more training data or to reduce the noise in the training data (remove fix data error, remove outliers).

16. What is test set and why would you want to use it?
The test set is the data you use to train your model. This way later on, you can validate your model with data it has never seen before.

17. What is the purpose of validation set?
The validation set is a set of data you use for validation purposes. This way, you can see how your model reacts to data its never seen before and compare your generalization error with the test data.

18. What can go wrong if you tune hyperparameters using the test set?
It can overfit the test set
data leakage

19. What is repeated cross-validation and why would you prefer it to using a single validation set?
It is a way of using multiple validation sets. Each model is evaluated once per validation set, after it is trained on the rest of the data. It helps to get more accurate measure of its performance.
